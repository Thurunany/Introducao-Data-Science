##Tabela de frequencia no software

```python
labels = ['1.50 - 1.57', '1.57 - 1.64', '1.64 - 1.70', '1.70 - 1.76', '1.76 - 1.83', '1.83 - 1.90']
```

```python
frequencia = pd.value_counts(
  pd.cut(
    x = x,
    bins = 6, # O numero de bins (intervalos) foi encontrado por uma função que o software disponibiliza.
    labels = labels,
  ),
  sort = False
)
frequencia
```

```python
percentual = pd.value_counts(
  pd.cut(
    x = x,
    bins = 6,
    labels = labels,
  ),
  sort = False,
  normalize = True
)
percentual
```

```python
tabela_frequencia = pd.DataFrame(
    {'Frequência': frequencia, 'Porcentagem (%)': percentual * 100}
)
tabela_frequencia
```
